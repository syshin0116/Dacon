```python
import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 13
plt.rcParams["figure.figsize"] = (10,10)
```

```python
# ArcGIS or GIS에서는 SHP파일만 가져와서 읽지만, python은 .dbf, .shx파일도 함께 있어야힌다. 
seoul_file = "./data/집계구 단위 서울 생활인구/통계지역경계(2016년+기준)/행정구역.shp"
seoul = gpd.read_file(seoul_file)
seoul
```

```python
ax = seoul.convex_hull.plot(color='purple', edgecolor="w")
ax.set_title("종로구 기초구역도")
ax.set_axis_off()
plt.show()
```
