```python
df_high_raw = pd.read_csv("서울시 고등학교 기본정보.csv", encoding="cp949")
df_high = df_high_raw[['학교명', '도로명주소']]
df_high['종류'] = '고등학교'
df_high.rename(columns = {'도로명주소' : '주소'}, inplace = True)
df_high
```
